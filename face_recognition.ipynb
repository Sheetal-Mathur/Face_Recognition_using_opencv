```python
import cv2
import numpy as np
import face_recognition
from google.colab.patches import cv2_imshow
```

```python
imgSRK=face_recognition.load_image_file("/content/srk.jpg")
imgSRK=cv2.cvtColor(imgSRK,cv2.COLOR_BGR2RGB)

imgSRK_test=face_recognition.load_image_file("/content/srktest.jpg")
imgSRK_test=cv2.cvtColor(imgSRK_test,cv2.COLOR_BGR2RGB)

faceloc=face_recognition.face_locations(imgSRK)[0]
encode1=face_recognition.face_encodings(imgSRK)[0]
cv2.rectangle(imgSRK,(faceloc[3],faceloc[0],faceloc[1],faceloc[2]),(255,0,0),2)

faceloc_test=face_recognition.face_locations(imgSRK_test)[0]
encode2=face_recognition.face_encodings(imgSRK_test)[0]
cv2.rectangle(imgSRK_test,(faceloc_test[3],faceloc_test[0],faceloc_test[1],faceloc_test[2]),(255,0,0),2)

results=face_recognition.compare_faces([encode1],encode2)
faceDis=face_recognition.face_distance([encode1],encode2)
cv2.putText(imgSRK_test,f'{results},{round(faceDis[0],2)}' ,(20,40),cv2.FONT_HERSHEY_SIMPLEX,1,(255,200,50),2,cv2.LINE_AA)
print(results,faceDis)

cv2_imshow(imgSRK)
cv2_imshow(imgSRK_test)
cv2.waitKey(0)
```
